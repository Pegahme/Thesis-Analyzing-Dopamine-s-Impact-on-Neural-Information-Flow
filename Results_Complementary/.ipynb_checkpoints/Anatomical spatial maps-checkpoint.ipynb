```python
import numpy as np
import os

output_dir = '/lustre/majlepy2/myproject/Anamaps_ICs'
labels_path = os.path.join(output_dir, 'aparc68_labels.txt')

# Load again for safety
import joblib
pca = joblib.load('/lustre/majlepy2/myproject/50_epochs_idtxl/group_pca_model.joblib')
ica_mixing = np.load('/lustre/majlepy2/myproject/50_epochs_idtxl/group_ica_mixing_matrix.npy')
with open(labels_path) as f:
    region_names = [line.strip() for line in f]

# IC spatial maps in region space: (regions, ICs)
anat_ic_maps = pca.components_.T @ ica_mixing  # (68, 23)

print("Anatomical IC maps shape:", anat_ic_maps.shape)  # should be (68, 23)

# Save as .npy and .csv
npy_path = os.path.join(output_dir, 'anatomical_ic_maps.npy')
csv_path = os.path.join(output_dir, 'anatomical_ic_maps.csv')
np.save(npy_path, anat_ic_maps)
np.savetxt(csv_path, anat_ic_maps, delimiter=',', header=','.join([f'IC{i+1}' for i in range(anat_ic_maps.shape[1])]), comments='')

# Also save region names again for convenience
with open(os.path.join(output_dir, 'anatomical_ic_map_regions.txt'), 'w') as f:
    for name in region_names:
        f.write(name + '\n')

print(f"Saved anatomical IC maps as .npy and .csv in {output_dir}")
```

```python
import numpy as np
import os
import matplotlib.pyplot as plt

output_dir = '/lustre/majlepy2/myproject/Anamaps_ICs'
anat_ic_maps = np.load(os.path.join(output_dir, 'anatomical_ic_maps.npy'))
with open(os.path.join(output_dir, 'aparc68_labels.txt')) as f:
    region_names = [line.strip() for line in f]

# Find top 5 regions (by absolute value) for each IC
top5_indices = []
for ic in range(anat_ic_maps.shape[1]):
    absvals = np.abs(anat_ic_maps[:, ic])
    top_inds = np.argsort(absvals)[-5:][::-1]  # top 5, descending
    top5_indices.extend(top_inds)

# Make unique and keep order
top5_indices = sorted(set(top5_indices), key=list(top5_indices).index)

# Slice the maps and region names
anat_ic_maps_top5 = anat_ic_maps[top5_indices, :]
region_names_top5 = [region_names[i] for i in top5_indices]

plt.figure(figsize=(14, 8))
im = plt.imshow(anat_ic_maps_top5, aspect='auto', cmap='RdBu_r')
plt.colorbar(im, label='Contribution')
plt.yticks(np.arange(len(region_names_top5)), region_names_top5)
plt.xticks(np.arange(anat_ic_maps.shape[1]), [f'IC{i+1}' for i in range(anat_ic_maps.shape[1])], rotation=90)
plt.title("Top 5 Contributing Anatomical Regions per IC (by absolute value)")
plt.tight_layout()

# Show interactively
plt.show()

# Save and overwrite
save_path = os.path.join(output_dir, 'anatomical_ic_maps_top5_heatmap.png')
plt.savefig(save_path, dpi=300)
plt.close()

print(f"Saved heatmap: {save_path}")
```

```python
import numpy as np
import os
import matplotlib.pyplot as plt

output_dir = '/lustre/majlepy2/myproject/Anamaps_ICs'
anat_ic_maps = np.load(os.path.join(output_dir, 'anatomical_ic_maps.npy'))
with open(os.path.join(output_dir, 'aparc68_labels.txt')) as f:
    region_names = [line.strip() for line in f]

n_regions, n_ics = anat_ic_maps.shape

plt.figure(figsize=(16, 10))
im = plt.imshow(anat_ic_maps, aspect='auto', cmap='RdBu_r', vmin=-np.max(np.abs(anat_ic_maps)), vmax=np.max(np.abs(anat_ic_maps)))
plt.colorbar(im, label='Contribution')
plt.yticks(np.arange(n_regions), region_names, fontsize=7)
plt.xticks(np.arange(n_ics), [f'IC{i+1}' for i in range(n_ics)], rotation=90)
plt.title("Annotated: Top 5 Contributing Anatomical Regions per IC (Full Matrix)")

# === Annotate top 5 per IC ===
for ic in range(n_ics):
    absvals = np.abs(anat_ic_maps[:, ic])
    top_inds = np.argsort(absvals)[-5:][::-1]  # indices of top 5 (by abs value)
    for region_idx in top_inds:
        plt.plot(ic, region_idx, 'ko', markersize=5, markeredgewidth=1.5, markerfacecolor='none')  # black circle

plt.tight_layout()

# Show interactively
plt.show()

# Save the annotated heatmap
save_path = os.path.join(output_dir, 'anatomical_ic_maps_annotated_top5_heatmap.png')
plt.savefig(save_path, dpi=300)
plt.close()

print(f"Saved annotated heatmap: {save_path}")
```

```python
import numpy as np
import os
import pandas as pd

output_dir = '/lustre/majlepy2/myproject/Anamaps_ICs'
anat_ic_maps = np.load(os.path.join(output_dir, 'anatomical_ic_maps.npy'))
with open(os.path.join(output_dir, 'aparc68_labels.txt')) as f:
    region_names = [line.strip() for line in f]

n_ics = anat_ic_maps.shape[1]

rows = []
for ic in range(n_ics):
    absvals = np.abs(anat_ic_maps[:, ic])
    top_inds = np.argsort(absvals)[-5:][::-1]
    top_regions = [region_names[i] for i in top_inds]
    row = {
        "IC": f"IC{ic+1}",
        "Top5_Region_Names": "; ".join(top_regions),
        "Top5_Region_Indices": "; ".join(map(str, top_inds)),
        "Top5_Contributions": "; ".join([f"{anat_ic_maps[i, ic]:.4f}" for i in top_inds])
    }
    rows.append(row)

df = pd.DataFrame(rows)
csv_path = os.path.join(output_dir, "top5_regions_per_IC.csv")
df.to_csv(csv_path, index=False)

print(f"Saved top 5 region report as CSV: {csv_path}")
print(df.head())
```

**Visualize an IC Spatial Map on the Brain**

```python
import numpy as np
import matplotlib.pyplot as plt
import os

output_dir = '/lustre/majlepy2/myproject/Anamaps_ICs'
anat_ic_maps = np.load(os.path.join(output_dir, 'anatomical_ic_maps.npy'))
with open(os.path.join(output_dir, 'aparc68_labels.txt')) as f:
    region_names = [line.strip() for line in f]

ic_idx = 0  # IC to plot
ic_weights = anat_ic_maps[:, ic_idx]

# Get top 10 regions (by absolute value)
N = 10
top_inds = np.argsort(np.abs(ic_weights))[-N:][::-1]
plt.figure(figsize=(7, 4))
plt.barh([region_names[i] for i in top_inds], ic_weights[top_inds], color='royalblue')
plt.xlabel('IC Weight')
plt.title(f"Top {N} Regions for IC{ic_idx+1}")
plt.tight_layout()
plt.savefig(os.path.join(output_dir, f"IC{ic_idx+1:02d}_top{N}_barplot.png"), dpi=200)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import os

output_dir = '/lustre/majlepy2/myproject/Anamaps_ICs'
anat_ic_maps = np.load(os.path.join(output_dir, 'anatomical_ic_maps.npy'))
with open(os.path.join(output_dir, 'aparc68_labels.txt')) as f:
    region_names = [line.strip() for line in f]

N = 10  # Number of top regions to show per IC

for ic_idx in range(anat_ic_maps.shape[1]):
    ic_weights = anat_ic_maps[:, ic_idx]
    # Find top-N regions by absolute value
    top_inds = np.argsort(np.abs(ic_weights))[-N:][::-1]
    plt.figure(figsize=(8, 5))
    bars = plt.barh([region_names[i] for i in top_inds], ic_weights[top_inds], color='royalblue')
    plt.xlabel('IC Weight')
    plt.title(f"Top {N} Regions for IC{ic_idx+1}")
    plt.tight_layout()
    #plt.show()
    save_path = os.path.join(output_dir, f"IC{ic_idx+1:02d}_top{N}_barplot.png")
    plt.savefig(save_path, dpi=200)
    plt.close()  # Don't display, just save

print(f"Saved {anat_ic_maps.shape[1]} barplots for all ICs in {output_dir}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import os

output_dir = '/lustre/majlepy2/myproject/Anamaps_ICs'
anat_ic_maps = np.load(os.path.join(output_dir, 'anatomical_ic_maps.npy'))  # (68, 23)
with open(os.path.join(output_dir, 'aparc68_labels.txt')) as f:
    region_names = [line.strip() for line in f]

num_ics = anat_ic_maps.shape[1]

for ic_idx in range(num_ics):
    ic_weights = anat_ic_maps[:, ic_idx]
    plt.figure(figsize=(8, 10))
    bars = plt.barh(range(len(region_names)), ic_weights, color='royalblue')
    plt.yticks(range(len(region_names)), region_names, fontsize=7)
    plt.xlabel('IC Weight')
    plt.title(f"All 68 Regions for IC{ic_idx+1}")
    plt.tight_layout()
    save_path = os.path.join(output_dir, f'IC{ic_idx+1:02d}_allregions_barplot.png')
    plt.savefig(save_path, dpi=200)
    plt.close()
    print(f"Saved: {save_path}")

print("Done: All IC all-region barplots saved.")
```
